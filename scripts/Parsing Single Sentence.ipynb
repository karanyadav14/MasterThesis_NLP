```python
import os
import sys
import csv
import requests
import pyconll
import stanza
from stanza.models.common.doc import Document
from stanza.utils.conll import CoNLL
```

## Converting to Conll Format

```python
#Given a sentence in a text file class should return its pos tagged file

class Parse_Single_Sentence:

  def __init__(self, path):
    self.filepath = path
    self.dirs = os.listdir(path)

  def sent2conll(self):

    for file in self.dirs:

        filename = self.filepath+str(file)

        if filename.endswith("_Input.txt"):
            filename_in = filename
            filename_out = filename_in.replace("_Input.txt", "_Conll.txt")

            
            with open(filename_in, 'r', encoding="utf-8") as file:
                

                result_list = []
                count = 1
                count_sent = 0
                for i in file:
                    sent = i.split(" ")
                    print(sent)
                    
                    for word in sent:
                      if word != " ":
                        result_list.append(str(count)+'\t'+word+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\t'+"_"+'\n')


                        if "।" in word:
                          count_sent +=1
                          count=1
                          result_list.append("\n")
                        else:
                            count+=1
                        
                  # print(result_list)
                print(count_sent)
                    
            with open(filename_out, 'w', encoding = 'utf-8') as f:
                for item in result_list:
                    f.write(item)
                
  def conll2pos(self):
      

    # for every file in the directory
    for file in self.dirs:
        # print(file)
        filename = self.filepath+str(file)

        if filename.endswith("_Conll.txt"):
            filename_in = filename
            filename_out = filename_in.replace("_Conll.txt", "_Pos.txt")
    
            print(filename_out)
            doc = CoNLL.conll2doc(filename_in)                    #Coverting txt file to stanza Document
            #print(type(doc))

            #Create Stanza POS tagging pipeline
            pos_pipeline = stanza.Pipeline(lang='hi', processors='tokenize,lemma, pos', tokenize_pretokenized=True)

            pos_tagged_doc = pos_pipeline(doc)
            
            CoNLL.write_doc2conll(pos_tagged_doc, filename_out)   

  def output(self):
    self.sent2conll()
    self.conll2pos()
```
