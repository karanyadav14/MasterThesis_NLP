```python
import os
import sys
import csv
import requests
import pyconll
from pyconll import util
from pyconll.util import find_ngrams as ngram
import stanza
from stanza.models.common.doc import Document
from stanza.utils.conll import CoNLL
from collections import Counter
```

```python
### Word Length


#giving the path of the directory
path = "/content/drive/MyDrive/HD Corpus/Files/Output_files/"
dirs = os.listdir(path)

# for every file in the directory
for file in dirs:
    filename = path+str(file)

    if filename.endswith("hi_1385_output.txt"):
        filename_in = filename
    
        
        file = pyconll.load_from_file(filename_in)

        
        #Unigram Frequency
        f_dict = Counter()
        
        word_len = {}
        
        for sentence in file:
            sent = (sentence.meta_value("Sentence")).split(" ")
#             print(sent)
            for i in sent:
                count = 0
                for j in i:
#                     print(j)
                    count+=1
                        
                word_len[i] = count
                
            print(word_len)
```
